## Step 3 Model training

```python
# drop the target 
X = df_cancer.drop( ['target'], axis = 1 )
```

```python
X
```

```python
y = df_cancer['target']
y
```

```python
# split data 
# We're going to use the testing data set which is data said that the model hasn't seen ever before
from sklearn.model_selection import train_test_split 
```

```python
# train = 80% test = 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
# use support vector 
from sklearn.svm import SVC
```

```python
# classification_report and confusion matrix 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC() # use support vector machine to do the classification
```

```python
# train data 
# use the fit method 
svc_model.fit(X_train, y_train)
```

## Step 4 Evalute model 

```python
y_predict = svc_model.predict(X_test)
```

```python
# confusion metrix 
# compare true value vs predict value  
cm = confusion_matrix(y_test,y_predict)
```

```python
# 
sns.heatmap(cm, annot = True)
```

## Step 5 Improve the model 

```python
# improve normalization
min_train = X_train.min()
```

```python
range_train = (X_train-min_train).max()
```

```python
X_train_scaled = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'],hue = y_train )
# training data without scaling
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'],hue = y_train )
# training data with scaling 
# we did unity normalization 
```

```python
min_test = X_test.min()
range_test = (X_test-min_train).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
svc_model.fit(X_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
cm =confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm,annot=True)
```

```python
# summary 
# accuracy
print( classification_report(y_test,y_predict) )
```

## IMPROVING THE MODEL - PART 2

```python
# optimize our C and gamma paramters of support vector machine
# the problem is how can we actualy search for the best values are the best parameter of C and gamma
# Skillern had method help us, search for parameter
```

```python
# program grids 
param_grid = { 'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001],'kernel':['rbf'] }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled,y_train)
# searching for best value of gamma and c 
```

```python
# get the best value of C and gamma
grid.best_params_
```

```python
# that kind of the output using the best or optimized value 
grid_prediction = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_prediction)
```

```python
sns.heatmap(cm, annot=True)
```

```python
# summary 
# accuracy
print( classification_report(y_test,grid_prediction) )
```
